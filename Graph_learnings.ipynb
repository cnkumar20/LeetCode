```python
def algo_for_creating_random_list_for_ad_directed_graph(number_of_edges,zero_mat):
    r,c = zero_mat.shape
    mat = zero_mat
    i=0
    collision = 0
    while(i < number_of_edges):
        r1 = randrange(r)
        c1 = randrange(c)
        if(mat[r1][c1]==1):
            collision +=1
            continue
        else:
            mat[r1][c1] = 1
            i +=1
    print(f"Collision: {collision}")
    return mat            
```

```python
#convert matrix to adjaceny list using node_label

def convert_directed(mat):
    dict_map = {}
    for label in node_label:
        dict_map[label] = set()
    for r in range(len(mat)):
        adj_list = dict_map[node_label[r]]
        for c in range(len(mat[0])):
            if(mat[r][c] == 1):
                adj_list.add(node_label[c])
    return dict_map

def convert_undirected(mat):
    dict_map = {}
    for label in node_label:
        dict_map[label] = set()
    for r in range(len(mat)):
        for c in range(len(mat[0])):
            if(mat[r][c] == 1):
                dict_map[node_label[r]].add(node_label[c])
                dict_map[node_label[c]].add(node_label[r])
    return dict_map

adj_list_directed = convert_directed(graph_matrix)
print(adj_list_directed)


            
                
    
```

```python
#create undirected graph
number_of_nodes = len(node_label)
max_number_of_edges = int(number_of_nodes*(number_of_nodes -1)/2)
number_of_edges = randrange(number_of_nodes,max_number_of_edges+1)
def algo_for_creating_random_list_for_ad_undirected_graph(number_of_edges,zero_mat):
    r,c = zero_mat.shape
    mat = zero_mat
    i=0
    collision = 0
    while(i < number_of_edges):
        r1 = randrange(r)
        c1 = randrange(c)
        if(c1==r1):
            collision += 1
            continue
        if(mat[r1][c1]==1 and mat[c1][r1]==1):
            collision +=1
            continue
        else:
            mat[r1][c1] = 1
            mat[c1][r1] = 1
            i +=1
    print(f"Collision: {collision}")
    return mat    
```

```python
#time to solve graph problems
# find distance between two arbitary nodes
from collections import deque

def find_dist(adj_list,src,dst):
    if(src==None):
        return
    queue = deque([src])
    visited = set()
    steps = 0
    while queue:
        l = len(queue)
        list_adj_nodes = list()
        while l >0:
            steps +=1
            node = queue.popleft()
            l -= 1
            if(node not in visited):
                visited.add(node)
            else:
                continue
            list_adj_nodes.extend(adj_list[node])
            for d in adj_list[node]:
                if(dst==d):
                    return steps
    return -1
        
        
        
    

```
